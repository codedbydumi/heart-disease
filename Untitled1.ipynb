


## üîç 1. `RandomizedSearchCV` vs `GridSearchCV`

Both are used for **hyperparameter tuning** (finding the best settings for your model).

* **GridSearchCV**

  * Tries **every possible combination** of hyperparameters you give.
  * Example: If you test `n_estimators = [100, 200]` and `max_depth = [5, 10]`, it will try all 4 combinations.
  * ‚úÖ More thorough
  * ‚ùå Slower, especially with many parameters

* **RandomizedSearchCV**

  * Instead of trying everything, it picks **random combinations** of hyperparameters (but still intelligently).
  * Example: If you allow 100 possible values, it might test only 20 randomly.
  * ‚úÖ Faster, works well for large search spaces
  * ‚ùå Might miss the absolute best one

üí° Analogy:

* GridSearch = ‚ÄúTry every dish on the menu to see the best.‚Äù
* RandomizedSearch = ‚ÄúPick random dishes, but still get something tasty faster.‚Äù

---

## üìä 2. `confusion_matrix`

* A table that shows **how well your classification model did**.
* It compares predictions vs actual labels.

Example for binary classification (Spam/Not Spam):

|                     | Predicted Spam      | Predicted Not Spam  |
| ------------------- | ------------------- | ------------------- |
| **Actual Spam**     | True Positive (TP)  | False Negative (FN) |
| **Actual Not Spam** | False Positive (FP) | True Negative (TN)  |

* **TP**: Correctly predicted spam
* **TN**: Correctly predicted not spam
* **FP**: Predicted spam but it wasn‚Äôt
* **FN**: Predicted not spam but it was

---

## üìù 3. `classification_report`

* A **summary of precision, recall, f1-score** for each class.
* Easier than manually calculating metrics.

Output looks like:

```
              precision   recall   f1-score   support
Class 0          0.85      0.90      0.87      100
Class 1          0.78      0.70      0.74       50
```

---

## üéØ 4. `precision_score`

* Out of all **predicted positives**, how many were correct?

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

‚úÖ Good when **false positives are costly**.
Example: Predicting if an email is spam ‚Üí we don‚Äôt want to wrongly classify important emails as spam.

---

## üì° 5. `recall_score`

* Out of all **actual positives**, how many did we correctly find?

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

‚úÖ Good when **false negatives are costly**.
Example: Cancer detection ‚Üí missing a positive case is very dangerous.

---

## ‚öñÔ∏è 6. `f1_score`

* The **balance between precision and recall**.
* Harmonic mean of precision & recall.

$$
F1 = 2 \cdot \frac{Precision \times Recall}{Precision + Recall}
$$

‚úÖ Useful when you need balance (like fraud detection).

---

## üìà 7. `RocCurveDisplay`

* Plots the **ROC Curve (Receiver Operating Characteristic Curve)**.
* Shows trade-off between **True Positive Rate (recall)** and **False Positive Rate** at different thresholds.
* Also gives **AUC (Area Under Curve)**:

  * 0.5 = random guessing
  * 1.0 = perfect model

üí° Example: In medical tests, you may adjust the threshold:

* Lower threshold ‚Üí catch more sick people (high recall) but risk more false alarms.
* Higher threshold ‚Üí fewer false alarms but miss some sick people.

## üîπ 1. Logistic Regression

‚û°Ô∏è Despite the name, **Logistic Regression is used for classification, not regression**.

* **What it does**:
  It learns the probability that something belongs to a class (e.g., spam or not spam).
  If the probability > 0.5 ‚Üí predict class 1, else ‚Üí class 0.

* **How it works (basic idea)**:

  * It fits a line (or curve in higher dimensions) that best separates the classes.
  * Uses the **sigmoid function (S-shaped curve)** to squash numbers between 0 and 1 ‚Üí probabilities.

* **When to use**:

  * Binary classification (yes/no, 0/1 problems).
  * Easy to interpret and fast.

‚úÖ Example: Predict whether a patient has diabetes or not.

---

## üîπ 2. K-Nearest Neighbors (KNN)

‚û°Ô∏è KNN is like "asking your neighbors for advice."

* **What it does**:
  For a new data point, it looks at the **k closest points** (neighbors) in the training data.

  * If most neighbors are Class A ‚Üí predict Class A.
  * If most neighbors are Class B ‚Üí predict Class B.

* **How it works (basic idea)**:

  * You choose a number `k` (like 3 or 5).
  * It uses distance (like Euclidean distance) to find the nearest neighbors.
  * Votes among them to decide the class.

* **When to use**:

  * Works well with small to medium datasets.
  * Not great with very large datasets (too slow).

‚úÖ Example: Predict whether a fruit is an apple or orange based on its weight and color by looking at its closest fruits.

---

## üîπ 3. Random Forest Classifier

‚û°Ô∏è Random Forest is an **ensemble model** (uses many models together).

* **What it does**:
  It builds **lots of decision trees** (like yes/no flowcharts) and combines their results.
  Each tree gives a "vote," and the forest chooses the majority vote.

* **How it works (basic idea)**:

  * Randomly selects subsets of data and features.
  * Builds many decision trees.
  * Averages the predictions (for regression) or takes majority vote (for classification).

* **Why it‚Äôs powerful**:

  * Handles missing values and outliers well.
  * Reduces overfitting compared to a single decision tree.

‚úÖ Example: Predict whether a loan will default based on income, credit score, and employment.

---

üìä **Quick comparison**

| Model               | Easy Explanation                 | Pros                       | Cons                   |
| ------------------- | -------------------------------- | -------------------------- | ---------------------- |
| Logistic Regression | Draws a line to separate classes | Fast, interpretable        | Only linear boundaries |
| KNN                 | Looks at neighbors‚Äô classes      | Simple, no training needed | Slow on large data     |
| Random Forest       | Many decision trees voting       | Very accurate, robust      | Harder to interpret    |

````python
# 1) Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, confusion_matrix, ConfusionMatrixDisplay,
                             roc_auc_score, RocCurveDisplay, average_precision_score,
                             PrecisionRecallDisplay)
import joblib

# 2) Load data
df = pd.read_csv("heart-disease.csv")

# If your columns differ, adjust these lists:
target_col = "target"
numeric_features = ["age","trestbps","chol","thalach","oldpeak","ca"]  # add/remove as per your file
categorical_features = ["sex","cp","fbs","restecg","exang","slope","thal"]  # adjust as needed

X = df.drop(columns=[target_col])
y = df[target_col].astype(int)

print("Class balance:", y.value_counts(normalize=True).round(3))

# 3) Train/valid split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 4) Preprocess: impute + scale numeric, impute + one-hot cat
numeric_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="median")),
    ("scale", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, numeric_features),
        ("cat", categorical_pipe, categorical_features)
    ]
)

# 5) Models
log_reg = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", LogisticRegression(max_iter=1000))
])

rf = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", RandomForestClassifier(random_state=42))
])

# 6) Hyperparameter grids
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

log_grid = {
    "clf__C": [0.1, 1.0, 3.0, 10.0],
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs"]
}

rf_grid = {
    "clf__n_estimators": [200, 400],
    "clf__max_depth": [None, 5, 10],
    "clf__min_samples_split": [2, 5],
    "clf__min_samples_leaf": [1, 2],
    "clf__class_weight": [None, "balanced"]
}

# 7) Tune with ROC-AUC (good for imbalanced data)
log_search = GridSearchCV(log_reg, log_grid, scoring="roc_auc", cv=cv, n_jobs=-1)
rf_search  = GridSearchCV(rf, rf_grid, scoring="roc_auc", cv=cv, n_jobs=-1)

log_search.fit(X_train, y_train)
rf_search.fit(X_train, y_train)

print("Best LOG params:", log_search.best_params_, "CV AUC:", round(log_search.best_score_, 4))
print("Best RF  params:", rf_search.best_params_,  "CV AUC:", round(rf_search.best_score_, 4))

# 8) Pick the better model on validation CV
best_search = rf_search if rf_search.best_score_ >= log_search.best_score_ else log_search
best_model = best_search.best_estimator_

# 9) Evaluate on held-out test set
y_proba = best_model.predict_proba(X_test)[:, 1]
y_pred = (y_proba >= 0.5).astype(int)

print("\nTest ROC-AUC:", round(roc_auc_score(y_test, y_proba), 4))
print("Test PR-AUC :", round(average_precision_score(y_test, y_proba), 4))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))

# Confusion matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred)).plot()
plt.title("Confusion Matrix (threshold=0.5)")
plt.show()

# ROC curve
RocCurveDisplay.from_predictions(y_test, y_proba)
plt.title("ROC Curve")
plt.show()

# PR curve
PrecisionRecallDisplay.from_predictions(y_test, y_proba)
plt.title("Precision-Recall Curve")
plt.show()

# 10) Simple model interpretation
# For Logistic Regression: show top coefficients
if isinstance(best_model.named_steps["clf"], LogisticRegression):
    clf = best_model.named_steps["clf"]
    ohe = best_model.named_steps["prep"].named_transformers_["cat"]["onehot"]
    cat_names = ohe.get_feature_names_out(categorical_features)
    feature_names = numeric_features + list(cat_names)
    coefs = pd.Series(clf.coef_[0], index=feature_names).sort_values(key=abs, ascending=False)
    print("\nTop LR coefficients:\n", coefs.head(15))

# For Random Forest: feature importances (after preprocessing)
if isinstance(best_model.named_steps["clf"], RandomForestClassifier):
    # Get feature names after preprocessing
    ohe = best_model.named_steps["prep"].named_transformers_["cat"]["onehot"]
    cat_names = ohe.get_feature_names_out(categorical_features)
    feature_names = numeric_features + list(cat_names)
    importances = pd.Series(best_model.named_steps["clf"].feature_importances_, index=feature_names)
    print("\nTop RF importances:\n", importances.sort_values(ascending=False).head(15))

# 11) Save final model
joblib.dump(best_model, "heart_disease_model.joblib")
print("\nSaved model to heart_disease_model.joblib")
````

# üß© What the models are

### 1. **Logistic Regression**

* **Type:** Linear model
* **Idea:** Finds a straight-line (or hyperplane) boundary in the feature space to separate classes.
* **How it works:** Calculates probabilities using the *sigmoid function* ‚Üí outputs values between 0 and 1, then applies a threshold (like 0.5).
* **Strengths:**

  * Simple, fast, interpretable (coefficients tell you feature influence).
  * Works well when the relationship between features and target is approximately **linear**.
* **Weaknesses:**

  * Struggles with complex, non-linear patterns.

---

### 2. **K-Nearest Neighbors (KNN)**

* **Type:** Instance-based / non-parametric model
* **Idea:** ‚ÄúA data point is most like its *neighbors*.‚Äù For prediction:

  * Look at the `k` closest training points.
  * Take majority vote (classification) or average (regression).
* **Strengths:**

  * Very intuitive, no training step (just stores data).
  * Captures local patterns well.
* **Weaknesses:**

  * Slow at prediction (must search neighbors).
  * Sensitive to irrelevant features and scaling.
  * Doesn‚Äôt extrapolate well.

---

### 3. **Random Forest**

* **Type:** Ensemble model (many decision trees)
* **Idea:**

  * Trains many trees on random subsets of data & features.
  * Each tree votes, and the majority decision is the final prediction.
* **Strengths:**

  * Handles non-linear patterns, interactions between features.
  * Resistant to overfitting compared to a single tree.
  * Works well out-of-the-box.
* **Weaknesses:**

  * Less interpretable than Logistic Regression.
  * Can be slower for very large datasets.

---

# ü•ä Why not just pick one model?

Because **no single algorithm is best for all problems** (this is called the **No Free Lunch theorem** in ML).

* Some datasets are simple and linearly separable ‚Üí Logistic Regression wins.
* Some have complex, wiggly boundaries ‚Üí Random Forest does better.
* Some are small and local-pattern heavy ‚Üí KNN can shine.

That‚Äôs why we usually **try multiple algorithms**, compare them with metrics (accuracy, F1, ROC-AUC, etc.), and then choose the best fit for the data.

In your project:

* Logistic Regression will tell you if the data is ‚Äúlinearly separable‚Äù and which features matter most.
* KNN tests if *local similarity* works for classification.
* Random Forest often gives the strongest baseline when patterns are complex.

* **Logistic Regression** ‚Üí a straight line dividing the plane.
* **KNN** ‚Üí lots of wiggly regions following the training data.
* **Random Forest** ‚Üí piecewise boxy regions formed by tree splits.

# üöÄ Takeaway

* We don‚Äôt trust just one model because each sees the data differently.
* By testing multiple models, we **learn more** about the data and pick the most reliable one.
* In the Heart Disease dataset, Logistic Regression is nice for interpretability (which risk factors matter), while Random Forest usually gives higher predictive performance.

# üîß What ‚ÄúTuning‚Äù Means

In ML, **tuning** means adjusting the *hyperparameters* of your model to find the settings that give the **best performance**.

* **Parameters** = learned from the data (e.g., the coefficients in Logistic Regression, or the splits inside a Decision Tree).
* **Hyperparameters** = you set them *before* training, they control how the model learns (e.g., `n_neighbors` in KNN, `C` in Logistic Regression, `n_estimators` in Random Forest).

‚û°Ô∏è **Tuning = searching for the best hyperparameters.**

---

# üéØ Why we tune?

Different hyperparameters can change how a model behaves:

* **KNN**: small `k` ‚Üí overfits, large `k` ‚Üí underfits.
* **Random Forest**: too few trees (`n_estimators`) ‚Üí unstable; too deep trees (`max_depth`) ‚Üí overfit.
* **Logistic Regression**: regularization strength (`C`) affects generalization.

If you don‚Äôt tune, you might end up with a model that looks good on training data but performs poorly on unseen test data.

---

# üîç Example (your KNN case)

* You tried different values of `n_neighbors = 1, 2, ‚Ä¶, 20`.
* For each, you measured accuracy on train and test sets.
* You‚Äôll pick the `k` that gives the highest **test accuracy** (or best balance of metrics).

That process = **hyperparameter tuning**.

---

# ‚öôÔ∏è Ways to Tune

1. **Manual tuning** (like your loop):
   Try different values, plot results, pick the best.
2. **Grid Search (GridSearchCV)**:
   Scikit-learn tries all combinations of hyperparameters with cross-validation.
3. **Random Search (RandomizedSearchCV)**:
   Randomly samples hyperparameters ‚Äî faster when the grid is huge.
4. **Bayesian / AutoML methods**:
   Smarter searches (like Optuna, Hyperopt, or Auto-sklearn).

---

‚úÖ So in short:
üëâ **Tuning = experimenting with hyperparameters until you find the sweet spot where your model generalizes best.**
